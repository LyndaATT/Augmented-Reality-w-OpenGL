```python
from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *
import sys

# Some api in the chain is translating the keystrokes to this binary string
# so instead of saying: ESCAPE = 27, we use the following.
ESCAPE = b'\x1b'

# Number of the glut window.
window = 0

# Rotations of arm parts
part_2 = 90.0
part_3 = 0.0
part_4 = 0.0
part_5 = 0.0
part_6 = 0.3

# length of part_4: upper arm
length_upper = 2.0
#param control translation 
delta = 0.0
# Main function
def main():

    global window
    
    # GLUT initialization
    glutInit(())
    
    # GLUT display mode
    glutInitDisplayMode(GLUT_DOUBLE | GLUT_RGB) 

    # Initial window size
    glutInitWindowSize(500,500)
    
    # Initial window position
    glutInitWindowPosition(100,100)
    
    # Window creation
    glutCreateWindow(b"Robot arm")

    # Initial colour and shading model
    glClearColor(0.0,0.0,0.0,0.0)
    glClearDepth(1.0)             # Enables Clearing Of The Depth Buffer
    glDepthFunc(GL_LESS)          # The Type Of Depth Test To Do
    glEnable(GL_DEPTH_TEST)       # Enables Depth Testing
    glShadeModel(GL_FLAT)         # Flat shading model 

    # Pass drawing function to glut
    glutDisplayFunc(draw)
    
    # When we are doing nothing, redraw the scene.
    glutIdleFunc(draw)
    
    # Register the function called when our window is resized.
    glutReshapeFunc(reshape)
    
    # Register the function called when the keyboard is pressed.
    glutKeyboardFunc(keys)

    # Start Event Processing Engine
    glutMainLoop()

# Draw robot arm parts
def draw():
    global part_2, part_3, part_4, part_5, part_6, length_upper,delta
    
    # Initial position
    glClear (GL_COLOR_BUFFER_BIT|GL_DEPTH_BUFFER_BIT)
    glPushMatrix()
    glTranslatef (0.0,-3.0,-7.0)
    
    # Part 1
    glColor3f(0.5,0.5,0.8)
    glPushMatrix()
    glScalef (4.0, 0.2, 4)
    glutWireCube (1.0)
    glPopMatrix()
    
    # Part 2
    glColor3f(0.5,0.8,0.5)
    glRotatef (part_2, 0.0, 0.0, 1.0)
    glTranslatef (1.0, 0.0, 0.0)
    glPushMatrix()
    glScalef (2.0, 0.2, 0.5)
    glutWireCube (1.0)
    glPopMatrix()

    # Part # 3
    glColor3f(0.5,0.8,0.8)
    glTranslatef (1.0, 0.0, 0.0)
    glRotatef (part_3, 0.0, 0.0, 0.8)
    glTranslatef (1.0, 0.0, 0.0)
    glPushMatrix()
    glScalef (2.0, 0.2, 0.5)
    glutWireCube (1.0)
    glPopMatrix()
    
    # Part # 4
    glColor3f(0.8,0.8,0.5)
    glTranslatef (1.0, 0.0, 0.0);
    glRotatef (part_4, 0.0, 1.0, 1.0);
    glTranslatef (1.0, 0.0, 0.0);
    glPushMatrix();
    glTranslatef (delta/2, 0.0, 0.0); #translate when stretching or reducing to avoid losing the position of the upper arm start
    glScalef (length_upper, 0.2, 0.5)
    glutWireCube (1.0)
    glPopMatrix()

    # Part # 5
    glColor3f(0.5,0.5,0.8)
    glTranslatef (1.0, 0.0, 0.0)
    glRotatef (part_5, 0.0, 0.0, 0.0)
    glTranslatef (0.3+delta, 0.0, 0.0) #translate lower arm while stretching or reducing upper arm to keep the coherent position

    glPushMatrix()
    glScalef (0.6, 1.5, 0.8)
    glutWireCube (1.0)
    glPopMatrix()
        
    # Part 6
    glColor3f(0.8,0.5,0.5)
    
    glTranslatef (0.0, 0.0, 0.0)
        
    glTranslatef (0.0,part_6, 0.0)
    glTranslatef (1.0, 0.0, 0.0)
    glPushMatrix()
    glScalef (1.5, 0.1, 0.1)
    glutWireCube (1.0)
    glPopMatrix()

    glTranslatef(0.0,-2*part_6,0.0)
    glTranslatef (0.0,0.0,0.0)
    glPushMatrix()
    glScalef (1.5, 0.1, 0.1)
    glutWireCube (1.0)
    glPopMatrix()

    glPopMatrix()
    glutSwapBuffers()

# Recalculate projection matrix if window size is modified
def reshape(w,h):
    glViewport (0, 0, w, h) # Reset The Current Viewport And Perspective Transformation
    glMatrixMode (GL_PROJECTION)
    glLoadIdentity()
    gluPerspective(65.0, float(w)/float(h), 1.0, 20.0)
    glMatrixMode(GL_MODELVIEW)
    glLoadIdentity()
    glTranslatef (0.0,0.0,-6.0)

# This function reads the keyboard to move the parts
def keys(*args):
    global part_2, part_3, part_4, part_5, part_6, length_upper, delta
    key = args[0]
    
    #rotating part 2
    if (key == b's'):
        if part_2<180.0:
            part_2 = (part_2+5.0) % 360.0
    elif (key==b'S'):
        if part_2>0.0:
            part_2 = (part_2-5.0) % 360.0
                    
   # 4. Modify this code so that you are able to stretch or reduce the length of the upper arm part with the keyboard keys "R" and "r" respectively
   #stretching/reducing length of part 2: upper arm
    
    elif (key==b'r'): #stretch
        if length_upper <= 5: #keep the arm in the screen
            length_upper += 0.05 #stretching by adding a factor to upper arm length
            delta += 0.05 #translating factor
    elif (key==b'R'): #reduce
        if delta > -2.0: #keep the upper arm over the middle arm
            length_upper -= 0.05 #reducing by substracting a factor to upper arm length
            delta -= 0.05 #translating factor
            
    #rotating part 3
    elif (key==b'e'):
        part_3 = (part_3+5.0) % 360.0
    elif (key==b'E'):
        part_3 = (part_3-5.0) % 360.0
    
    #rotating part 4
    elif (key==b'w'):
        part_4 = (part_4+5.0) % 360.0
    elif (key==b'W'):
        part_4 = (part_4-5.0) % 360.0

    #rotating part 6
    elif (key==b'f'):
        if part_6>0.1:
            part_6 = (part_6-0.05)
    elif (key==b'F'):
        if part_6<0.7:
            part_6 = (part_6+0.05)
            
    #exiting the simulation
    elif (key == ESCAPE):
        glutLeaveMainLoop()
        return  

# Print message to console, and kick off the main to get it rolling.
print("Hit ESC key to quit.")
main()
```

```python
from OpenGL.GL import *
from OpenGL.GLU import *
from OpenGL.GLUT import *
import sys

# Some api in the chain is translating the keystrokes to this binary string
# so instead of saying: ESCAPE = 27, we use the following.
ESCAPE = b'\x1b'

# Number of the glut window.
window = 0

# Rotations of arm parts
arm_1 = 0.0
arm_2 = 20.0
hand = 0.0
finger = 0.5
length_upper, delta = 4.0,0.0

# Main function
def main():

    global window
    
    # GLUT initialization
    glutInit(())
    
    # GLUT display mode
    glutInitDisplayMode(GLUT_DOUBLE | GLUT_RGB) 

    # Initial window size
    glutInitWindowSize(500,500)
    
    # Initial window position
    glutInitWindowPosition(100,100)
    
    # Window creation
    glutCreateWindow(b"Robot arm")

    # Initial colour and shading model
    glClearColor(0.0,0.0,0.0,0.0)
    glClearDepth(1.0)             # Enables Clearing Of The Depth Buffer
    glDepthFunc(GL_LESS)          # The Type Of Depth Test To Do
    glEnable(GL_DEPTH_TEST)       # Enables Depth Testing
    glShadeModel(GL_FLAT)         # Flat shading model 

    # Pass drawing function to glut
    glutDisplayFunc(draw)
    
    # When we are doing nothing, redraw the scene.
    glutIdleFunc(draw)
    
    # Register the function called when our window is resized.
    glutReshapeFunc(reshape)
    
    # Register the function called when the keyboard is pressed.
    glutKeyboardFunc(keys)

    # Start Event Processing Engine
    glutMainLoop()

    
# Draw robot arm parts
def draw():
    global arm_1 ,arm_2, hand, finger,length_upper, delta
    # Initial position
    glClear (GL_COLOR_BUFFER_BIT|GL_DEPTH_BUFFER_BIT)
    glPushMatrix()
    glTranslatef (0.0,-3.0,-7.0)

    # lower arm
    glColor3f(0.8,0.8,0.8)
    glTranslatef (-1.0, 0.0, 0.0)
    glRotatef (arm_1, 0.0, 0.0, 0.8)
    glTranslatef (1.0, 0.0, 0.0)
    glPushMatrix()
    glScalef (4.5, 1.5, 0.5)
    glutSolidCube (1.0)
    glPopMatrix()
    
    # upper arm
    glColor3f(0.8,0.8,0.8)
    glTranslatef (2.2, 0.0, 0.0);
    glRotatef (arm_2, 0.0, 1.0, 1.0);
    glTranslatef (2.0, 0.0, 0.0);
    glPushMatrix();
    glTranslatef (delta/2, 0.0, 0.0)
    glScalef (length_upper, 1.5, 0.5)
    glutSolidCube(1.0)
    glPopMatrix()


    # hand
    glColor3f(0.8,0.5,0.5)
    glTranslatef (2.5, 0.0, 0.0)
    glRotatef (hand, 0.0, 0.0, 0.7)
    glTranslatef (0.3+delta, 0.0, 0.0)
    glPushMatrix()
    glScalef (1.5, 1.5, 0.8)
    glutSolidCube(1.0)
    glPopMatrix()

        
    
    # Fingers: 
    
    glColor3f(0.8,0.5,0.5)      
    glTranslatef (0.0,finger, 0.0)
    glTranslatef (1.0, 0.0, 0.0)
    glPushMatrix()
    glScalef (1.0, 0.1, 0.1)
    glutSolidCube (1.5)
    glPopMatrix()
    
    glColor3f(0.8,0.5,0.5)
    glTranslatef (-1.0, -0.8, 0.0)
    glTranslatef (0.0,finger, 0.0)
    glTranslatef (1.0, 0.0, 0.0)
    glPushMatrix()
    glScalef (1.3, 0.1, 0.1)
    glutSolidCube (1.5)
    glPopMatrix()
   
    glColor3f(0.8,0.5,0.5)
    glTranslatef (-1.0, -0.8, 0.0)
    glTranslatef (0.0,finger, 0.0)
    glTranslatef (1.0, 0.0, 0.0)
    glPushMatrix()
    glScalef (1.6, 0.1, 0.1)
    glutSolidCube (1.5)
    glPopMatrix()
    
    glColor3f(0.8,0.5,0.5)
    glTranslatef (-1.0, -0.8, 0.0)
    glTranslatef (0.0,finger, 0.0)
    glTranslatef (1.0, 0.0, 0.0)
    glPushMatrix()
    glScalef (1.4, 0.1, 0.1)
    glutSolidCube (1.5)
    glPopMatrix()

    glColor3f(0.8,0.5,0.5)
    glTranslatef (-1.0, -0.8, 0.0)
    glTranslatef (0.0,finger, 0.0)
    glTranslatef (1.0, 0.0, 0.0)
    glPushMatrix()
    glScalef (1.0, 0.1, 0.1)
    glutSolidCube (1.5)
    glPopMatrix()
    
    
    glPopMatrix()
    glutSwapBuffers()

# Recalculate projection matrix if window size is modified
def reshape(w,h):
    glViewport (0, 0, w, h) # Reset The Current Viewport And Perspective Transformation
    glMatrixMode (GL_PROJECTION)
    glLoadIdentity()
    gluPerspective(65.0, float(w)/float(h), 1.0, 20.0)
    glMatrixMode(GL_MODELVIEW)
    glLoadIdentity()
    glTranslatef (-1.0,0.0,-5.0)

# This function reads the keyboard to move the parts
def keys(*args):
    global arm_1 ,arm_2, hand, finger,length_upper, delta
    key = args[0]
   
    #rotating lower arm
    if (key == b'a'):
        if arm_1<180.0:
            arm_1 = (arm_1+5.0) % 360.0
    elif (key==b'A'):
        if arm_1>0.0:
            arm_1 = (arm_1-5.0) % 360.0
            
    #rotating upper arm
    elif (key==b'b'):
        arm_2 = (arm_2+5.0) % 360.0
    elif (key==b'B'):
        arm_2 = (arm_2-5.0) % 360.0

    #stretching/reducing length of arm_2: upper arm
    elif (key==b'r'): #stretch
        if length_upper <= 5: #keep the arm in the screen
            length_upper += 0.05 #stretching by adding a factor to upper arm length
            delta += 0.05 #translating factor
    elif (key==b'R'): #reduce
        if delta > -4.0: #keep the upper arm over the middle arm
            length_upper -= 0.05 #reducing by substracting a factor to upper arm length
            delta -= 0.05 #translating factor
            
    #rotating hand
    elif (key==b'c'):
        hand = (hand+5.0) % 360.0
    elif (key==b'C'):
        hand = (hand-5.0) % 360.0
    

    #exiting the simulation
    if (key == ESCAPE):
        glutLeaveMainLoop()
        return  

# Print message to console, and kick off the main to get it rolling.
print("Hit ESC key to quit.")
main()
```
